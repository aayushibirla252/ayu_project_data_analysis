# EXPLORATORY DATA ANALYSIS

# Calculating  overall sales trend

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
df['month_year']=df['order_date'].apply(lambda x:x.strftime('%Y-%m'))
```

```python
#grouping month year
df_trend =df.groupby('month_year').sum()['sales'].reset_index()
```

```python
#setting the figure size
plt.figure(figsize=(16,7))
#tilt for 90 degree
plt.xticks(rotation='vertical',size=9)

plt.plot(df_trend['month_year'],df_trend['sales'])
```

# TOP 10 PRODUCT BY SALES


```python
#grouping product name column
prod_sales=pd.DataFrame(df.groupby('product_name').sum()['sales'])
```

```python
#sorting prod_sales column in descending order
prod_sales=prod_sales.sort_values('sales',ascending=False)
```

```python
#Finding top 10 peoduct by sales
prod_sales[:10]
```

# MOST SELLING PRODUCTS

```python
#Grouping product name
most_selling_product=pd.DataFrame(df.groupby('product_name').sum()['quantity'])
```

```python
#Sorting in descending format
most_selling_product=most_selling_product.sort_values('quantity',ascending=False)
```

```python
#fetching most selling products
most_selling_product[:10]
```

# MOST PREFERRED SHIP MODE

```python
#changing figure size
plt.figure(figsize=(11,7))
#using seaborn library for visualisation

sns.countplot(df['ship_mode'])
```
